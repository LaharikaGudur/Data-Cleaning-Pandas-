```python
## importing Libraries
import pandas as pd
import numpy as np
```

```python
# Read the dataset in csv format
data = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
print(data)
```

```python
data.head()
data.isnull().sum()
```

```python
# deleting index column.
data.drop(columns = ["Index"],inplace=True)

print(data)
```

```python
# replace the NaN values in the age column with mean value.
mean_age = data.Age.mean()
data['Age'] = data['Age'].fillna(mean_age)
data['Age'] = data['Age'].round(decimals = 1)
print(data)
```

```python
#Converting the age in integer format
data['Age'] = data['Age'].astype(int)
data.head()
```

```python
# Removing $ sign from Salary column
data['Salary'] = data['Salary'].apply(lambda x:x.replace('$',''))
data.head()

data['Salary'] = data['Salary'].apply(lambda y:y.replace('k','000'))
data.head()
```

```python
# Replace -1 from Rating column with NaN value.
data["Rating"] = data["Rating"].replace(-1,np.nan)

# Replace NaN values with average rating value
data['Rating']=data["Rating"].fillna(data['Rating'].mean()).round(decimals=1)
print(data)
```

```python
# clean data for location method 1
data['Location'] = data['Location'].replace('Australia Aus','Australia, Aus')
data['Location'] = data['Location'].replace('India In','India,In')
print(data)
```

```python
# replace -1 value with TRUE
data['Easy Apply'] = data['Easy Apply'].replace("-1","TRUE")

print(data)
```

```python
# we cannot replace the -1 value of established column with its mean or median value .so simply replace -1 with unknown.
data['Established'] = data['Established'].replace(-1,np.nan)
data['Established'] = data['Established'].fillna("Unknown")
print(data)
```

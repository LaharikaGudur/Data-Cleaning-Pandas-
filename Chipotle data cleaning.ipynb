```python
import pandas as pd
import numpy as np
```

```python
# reading tsv file
data = pd.read_csv("/content/chipotle.tsv",sep='\t')
data.head(20)
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data['item_price'] = data['item_price'].apply(lambda x:x.replace('$',''))
data['item_price'] = data['item_price'].astype(float)
data.dtypes
data.head(20)

```

```python
#Removing special character(-) from item_name column
data['item_name'] = data['item_name'].str.replace("Chips and Tomatillo-Green Chili Salsa","Chips and Tomatillo Green Chili Salsa")
data.head(20)
```

```python
# Replace the NaN values from choice_Description column with 'no choice' because here we can't add any values like mean median or mode.
data['choice_description'] = data['choice_description'].fillna('no choice')

# Remove special characters from choice_ description column
data['choice_description'] = data['choice_description'].str.replace("Tomatillo-Red Chili Salsa","Tomatillo Red Chili Salsa")
data['choice_description'] = data['choice_description'].str.replace("Tomatillo-Green Chili Salsa","Tomatillo Green Chili Salsa")
data.head(20)
```

```python
# Check irregularities in order id
irregularity = data['order_id'].value_counts()
print(irregularity)

# Check the order id is sequential or not
data['order_id'].astype(int).is_monotonic_increasing
```
